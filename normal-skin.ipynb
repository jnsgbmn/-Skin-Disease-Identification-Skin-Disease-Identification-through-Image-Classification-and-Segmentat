```python
from PIL import Image
import os.path, sys
import random

path = "./normal"
dirs = os.listdir(path)

random_box = [(70,5,120,45),(30,70,80,120),(120,70,160,120)]

def crop():
    for item in dirs:
        fullpath = os.path.join(path, item)
        if os.path.isfile(fullpath):
            im = Image.open(fullpath)
            f, e = os.path.splitext(fullpath)
            width, height = im.size
            for i in range(0,3):
                box = random_box[i]
                imCrop = im.crop(box)
                imCrop = imCrop.resize((200,200), Image.NEAREST)
                imCrop.save(f + str(i) + 'Cropped.bmp',"BMP", quality=100)
            os.remove(fullpath)
            print(width, height)

crop()
```
